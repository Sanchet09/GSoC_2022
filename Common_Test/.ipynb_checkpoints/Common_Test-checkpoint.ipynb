## Configurations

```python
DIR = 'D:\sanchet acer backup\BITS\GSoC\dataset\dataset\\'
TRAIN_DIR = DIR + 'train\\'
VAL_DIR = DIR + 'val\\'

TYPES = ['no','sphere','vort']

DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
print(DEVICE)

BATCH_SIZE = 16

LR = 0.001
EPOCHS = 100

MODELS = ['tresnet_l','efficientnet_b0','mobilenetv3_large_100','inception_v3'] #check the names once
```

## Image Augmentation

```python
train_augs = A.Compose(
    [
        A.SmallestMaxSize(max_size=150),
        A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=360, p=0.5),
        A.RandomCrop(height=128, width=128),
        # A.RGBShift(r_shift_limit=15, g_shift_limit=15, b_shift_limit=15, p=0.5),
        # A.RandomBrightnessContrast(p=0.5),
        A.MultiplicativeNoise(multiplier=(0.9,1.1), per_channel=True, p=0.2),
        A.Normalize(mean=0.485, std=0.224),
        # A.HueSaturationValue(hue_shift_limit=0.2, sat_shift_limit=0.2, val_shift_limit=0.2, p=0.5),
        A.RandomBrightnessContrast(brightness_limit=(-0.2,0.2), contrast_limit=(-0.2, 0.2), p=0.2),
        ToTensorV2(),
    ]
)

test_augs = A.Compose(
    [
        A.SmallestMaxSize(max_size=150),
        A.CenterCrop(height=128, width=128),
        A.Normalize(mean=0.485, std=0.224),
        ToTensorV2(),
    ]
)
```

## Create train, valid and test sets

```python
train_image_paths = []
classes = []

#1.
#create the train_image_paths and classes for storing the class labels for classification
for root,dirs,files in os.walk(TRAIN_DIR):
    for name in dirs:
        classes.append(name)
    for name in files:
        train_image_paths.append(os.path.join(root,name))

random.shuffle(train_image_paths)

# print(train_image_paths)
print('train_image_path example: ', train_image_paths[0])
print('class example: ', classes[0])
print(classes)

#2.
# split train valid from train paths (80,20)
train_image_paths, valid_image_paths = train_image_paths[:int(0.8*len(train_image_paths))], train_image_paths[int(0.8*len(train_image_paths)):] 

#3.
# create the test_image_paths
test_image_paths = []
for root,dirs,files in os.walk(VAL_DIR):
    for name in files:
        test_image_paths.append(os.path.join(root,name))

print("Train size: {}\nValid size: {}\nTest size: {}".format(len(train_image_paths), len(valid_image_paths), len(test_image_paths)))
```

## Create the index-to-class and class-to-index mappings

```python
# Since we cannot use the class labels directly in classification models, we create their mappings to index and index-to-class
index_to_class = {i:j for i,j in enumerate(classes)}
class_to_index = {value:key for key,value in index_to_class.items()}
class_to_index
```

## Create Custom Dataset

```python
class DarkMatterSubstructure(Dataset):
    
    # __init__ function is executed when an instance of the DarkMatterSubstructure class is created.
    # It contains the operations which are to be executed on the entire set as a whole when the class is instantiated
    def __init__(self, image_paths, augmentations=None):
        self.image_paths = image_paths
        self.augmentations = augmentations
    
    # __len__ function returns the length of the dataset to be used later by the DataLoader
    def __len__(self):
        return len(self.image_paths)

    # __getitem__ function returns one data point at a time which is sampled and returned
    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        image = np.load(image_path)
        image = image.squeeze()
        image = image.astype(np.float32)
        # image = 255 * image
        image = np.stack((image,)*3,axis=-1)
        
        label = image_path.split('\\')[-2]
        label = class_to_index[label]

        if self.augmentations is not None:
            image = self.augmentations(image=image)['image']

        return image,label
    

# create datasets:
train_dataset = DarkMatterSubstructure(train_image_paths, train_augs)
valid_dataset = DarkMatterSubstructure(valid_image_paths, test_augs)
test_dataset = DarkMatterSubstructure(test_image_paths, test_augs)
```

## Visualize images after Augmentation

```python
def visualize_augmentations(dataset, idx = 0, samples = 10, cols = 5, random_img = False):
    dataset = copy.deepcopy(dataset)
    dataset.augmentations = A.Compose([t for t in dataset.augmentations if not isinstance(t, (A.Normalize, ToTensorV2))])
    # dataset.augmentations = None
    rows = samples // cols
    
        
    figure, ax = plt.subplots(nrows=rows, ncols=cols, figsize=(12, 8))
    for i in range(samples):
        if random_img is True:
            idx = np.random.randint(1,len(train_image_paths))
        img, label = dataset[idx]
        ax.ravel()[i].imshow(img)
        ax.ravel()[i].set_axis_off()
        ax.ravel()[i].set_title(index_to_class[label])
        # print(img.shape)
    plt.tight_layout(pad=1)
    plt.show()    

visualize_augmentations(train_dataset,np.random.randint(1,len(train_image_paths)), random_img = True)
```

## Creating the DataLoader 

```python
train_loader = DataLoader(train_dataset, BATCH_SIZE, shuffle = True)
valid_loader = DataLoader(valid_dataset, BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_augs, BATCH_SIZE, shuffle = True)
```

## Create Model

```python
class DarkMatterModel(nn.Module):
    def __init__(self):
        super(DarkMatterModel, self).__init__()
        self.backbone = timm.create_model(MODELS[0], pretrained = True, num_classes = 3)
    
    def forward(self, image, label = None):
        output = self.backbone(image)
        if label is not None:
            loss = nn.CrossEntropyLoss()(output, label)
            return output, loss
        return output

#create the DarkMatterModel instance
model = DarkMatterModel()
model.to(DEVICE)
```

## Create Train and Eval Function

```python
def train_fn(model, dataloader, optimizer):
    total_loss = 0.0
    model.train() #dropout ON

    for data in tqdm(dataloader):
        image, label = data
        image, label = image.to(DEVICE), label.to(DEVICE)

        output, loss = model(image, label=label)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(dataloader)
```

```python
def eval_fn(model, dataloader):
    total_loss = 0.0
    model.eval() #dropout OFF

    with torch.no_grad():
        for data in tqdm(dataloader):
            image, label = data
            image, label = image.to(DEVICE), label.to(DEVICE)

            output, loss = model(image, label=label)

            total_loss += loss.item()
        
        return total_loss / len(dataloader)
```

## Training

```python
optimizer = torch.optim.Adam(model.parameters(), lr = LR)
```

```python
best_valid_loss = np.Inf

for i in range(EPOCHS):
    train_loss = train_fn(model, train_loader, optimizer)
    valid_loss = eval_fn(model, valid_loader)

    if valid_loss < best_valid_loss:
        torch.save(model.state_dict(), 'best_model.pt')
        print('Weights are saved')

        best_valid_loss = valid_loss

    print(f'Epoch:{i+1} Train loss:{train_loss}  Valid_loss:{valid_loss}')

# might need to import in place batch normalization libraries so execute: pip install git+https://github.com/mapillary/inplace_abn.git
# for python setup.py not working error on running the above command try the following:
#   -> pip install ez_setup
#   -> download the CUDA toolkit according to system configurations and install it
```
